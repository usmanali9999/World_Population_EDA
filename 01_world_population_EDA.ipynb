```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('world_population.csv')
```

```python
df.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)     # display  floating_nums with 2 decimal places
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()    # Counting all Null Values
```

```python
df.nunique()         #  Count of Unique Values
```

```python
# Sorting on Values Descending
df.sort_values(by="2022 Population" , ascending=False)
```

```python
df.corr(numeric_only=True)
```

```python
# Correlation Heatmap between Columns using Seaborn

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.rcParams['figure.figsize'] = (15, 9)

plt.show()
```

```python
df2 = df.groupby('Continent')[['1970 Population',
                                '1980 Population',
                                '1990 Population',
                                '2000 Population',
                                '2010 Population',
                                '2015 Population',
                                '2020 Population',
                                '2022 Population']].mean().sort_values(by="2022 Population", ascending=False)

df2
```

```python
df3 = df2.transpose()
```

```python
df3
```

```python
plt.figure(figsize=(5, 3))
df3.plot()

plt.xlabel("Year")
plt.ylabel("Average Population")
plt.title("Average Population Growth by Continent")
plt.xticks(rotation=45)

plt.show()
```
